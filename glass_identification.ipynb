# Model Building

```python
import itertools
from numpy import sort
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split, KFold, cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from xgboost import XGBClassifier, plot_tree, plot_importance
```

```python
# Train test split
x= df.drop(columns= ['Type'])
y= df['Type']
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state= 0)
```

```python
xg= XGBClassifier()
xg.fit(x_train, y_train)
```

```python
plot_tree(xg, num_trees=0, rankdir='LR')
fig=plt.gcf()
fig.set_size_inches(100,50)
```

```python
plot_tree(xg, num_trees=5, rankdir='LR')
fig=plt.gcf()
fig.set_size_inches(100,50)
```

```python
y_pred= xg.predict(x_test)
accuracy= accuracy_score(y_test, y_pred)
print('Accuracy:-', accuracy*100)
```

```python
cm=confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
plot_importance(xg)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
feature_importance= sort(xg.feature_importances_)

for each_thresh in feature_importance:
    selection= SelectFromModel(xg, threshold= each_thresh, prefit= True)
    modified_x_train= selection.transform(x_train)
    selection_model= XGBClassifier()
    selection_model.fit(modified_x_train, y_train)
    modified_x_test= selection.transform(x_test)
    prediction= selection_model.predict(modified_x_test)
    accuracy= accuracy_score(y_test, prediction)
    print('Threshold:- ', each_thresh, 'Number of Features:-', modified_x_train.shape[1], 'Model Accuracy:-', accuracy*100 )
```

```python
strat= StratifiedKFold(n_splits=40, random_state=0)
xg_strat= XGBClassifier()
sf_results= cross_val_score(xg_strat, x_train, y_train, cv= strat, verbose=True)
print('Accuracy:-', sf_results.mean()*100)
```
